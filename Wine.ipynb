##### calculating distribution of this features

```python
plt.figure(figsize=(11,6))

sns.histplot(data=data, x='fixed acidity', color='orange', alpha=0.9, bins =5)

plt.title('Distribution of Fixed Acidity')
plt.xlabel("Fixed Acidity")
plt.ylabel('Count')
plt.show()
```

##### In this graph we can observ that data are normally disturduted and most of the values lies between 6 and 8.

##### Applying mean and median to the graph

```python
plt.figure(figsize=(11,6))

sns.histplot(data=data, x='fixed acidity', color='pink', alpha=0.9, bins =5)

plt.title('Distribution of Fixed Acidity')
plt.xlabel("Fixed Acidity")
plt.ylabel('Count')
plt.vlines(x=data['fixed acidity'].mean(), ymin=0, ymax=4000, color='blue', alpha=0.7, label='Mean')
plt.vlines(x=data['fixed acidity'].median(), ymin=0, ymax=4000, color='red', alpha=0.7, label='Median')
plt.legend()
plt.show()
```

##### We can conclude that the man and median are very close to each other

##### Applying graph for valotile acidity

```python
plt.figure(figsize=(11,6))

sns.histplot(data=data, x='volatile acidity', color='orange', alpha=0.9, bins =5)

plt.title('Distribution of Volatile Acidity')
plt.xlabel("Volatile Acidity")
plt.ylabel('Count')
plt.show()
```

##### We can conculde thta this graph is positively skwed so we have to plot displot to understand the disturbution

```python
plt.figure(figsize=(11,6))

sns.distplot(data['volatile acidity'], color='skyblue')

plt.title('Distribution of Volatile Acidity')
plt.xlabel("Volatile Acidity")
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(11,6))

sns.distplot(data['volatile acidity'], color='skyblue')

plt.title('Distribution of Volatile Acidity')
plt.xlabel("Volatile Acidity")
plt.ylabel('Density')
plt.vlines(data['volatile acidity'].mean(), ymin=0, ymax=5, color='green', label='Mean')
plt.vlines(data['volatile acidity'].median(), ymin=0, ymax=5, color='black', label='Median')
plt.legend()
plt.show()
```

##### The plot above shows the normal distribution
##### Mean and median are close to each other
