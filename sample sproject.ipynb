# Finding most loyal customer

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
customer_stats = df.groupby("Customer Name").agg({"Order Date": "nunique","Sales":"sum", "Discount":"mean"}).reset_index()
```

```python
customer_stats.rename(columns={"Order Date":"Order_Frequency","Sales":"Total_Sales","Profit":"Total_Profit","Discount":"Avg_Discount"}, inplace=True)
```

```python
loyal_customers = customer_stats.sort_values(by=["Order_Frequency","Total_Sales"],ascending=[False,False]).head(10)
loyal_customers
```

# finding out on which day did the Business flourish

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
profit_by_date = df.groupby("Order Date")["Profit"].sum().reset_index()
```

```python
max_profit_date = profit_by_date.loc[profit_by_date["Profit"].idxmax()]
```

```python
print("Date with highest profit:",max_profit_date["Order Date"])
print("Profit amount:", max_profit_date["Profit"])
```
