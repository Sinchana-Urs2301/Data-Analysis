# Visual representation of Hospital Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/sinch/Downloads/Hospital ER_Data.csv" ,  encoding="ISO-8859-1")
data
```

# Disturdution of Patients Gender

```python
sns.countplot(x = 'Patient Gender' , data = data)
plt.title('Patient disturbution by Gender')
plt.show()
```

# Finding Patients Visited the ER by gender

```python
gender_counts = data['Patient Gender'].value_counts()
print(gender_counts)
```

# Graphical Represenation of highest number of refrerals received by Departments

```python
dept_counts = data['Department Referral'].value_counts()
dept_counts.plot(kind='bar')
plt.xlabel('Department')
plt.ylabel('Number of Patients')
plt.show()
```

# Finding which Department receives the highest number of refferals

```python
dept_counts = data['Department Referral'].value_counts()
print(dept_counts)
```

```python
dept_counts.plot(kind='bar')
plt.title('Patients by Department Referral')
plt.xlabel('Department')
plt.ylabel('Number of Patients')
plt.show()
```
