### Finding correlation between two variables by visualizations

```python
plt.figure(figsize= (10,4))
```

```python
sns.scatterplot(data=data , x='healthy_eating', y='active_lifestyle', color='grey', alpha=0.8)
plt.title('Healthy Eating   vs   Active Lifestyle')
plt.xlabel('Healthy Eating')
plt.ylabel('Active Lifestyle')
plt.show()
```

```python
sns.scatterplot(data=data , x='salary', y='healthy_eating', color='red', alpha=0.8)
plt.title('Salary   vs   Healthy Eating ')
plt.xlabel('Salary')
plt.ylabel('Healthy Eating')
plt.show()
```

```python
sns.scatterplot(data=data , x='age', y='salary', color='blue', alpha=0.8)
plt.title('Age   vs   Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

##### From this three plots we can conclude that healthy eating affects the active lifestyle in the same way salary affects haelthy eating and age with salary

##### Ploting graph for counting purpose

```python
plt.figure(figsize=(10,5))

sns.histplot(data=data, x='salary', color='orange', alpha=0.4)
plt.title('Count of Salary')
plt.xlabel('Salary')
plt.ylabel('Count')
plt.show()
```

##### We can conclude from this graph that the salary is not distributed equally.
##### The least as more number of employee and high salary as less number of employees
